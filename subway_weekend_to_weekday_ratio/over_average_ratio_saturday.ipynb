```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/total_getoff.csv')
df.head()
```

```python
dow = ['Friday', 'Saturday', 'Sunday']

weekday = df.loc[~df['요일'].isin(dow)]
weekday.head()
```

```python
#주말을 토요일만
weekend = df.loc[df['요일']=='Saturday']
```

```python
weekday = pd.DataFrame(weekday.groupby(['역명'], as_index=False).mean())
weekday
```

```python
weekend = pd.DataFrame(weekend.groupby(['역명'], as_index=False).mean())
weekend.head()
```

```python
#주중 대비 주말 비율 구하기
weekday_values = weekday.iloc[:, 1:].values
weekend_values = weekend.iloc[:, 1:].values

weekend_ratio = np.round((weekend_values/weekday_values)*100,2)
weekend_ratio
```

```python
ratio = weekend.copy()
```

```python
ratio.iloc[:, 1:] = weekend_ratio
ratio
```

```python
weekend_copy = ratio.copy()
```

```python
weekend_copy['평균'] = '평균'
weekend_copy.head()
```

```python
average = pd.DataFrame(weekend_copy.groupby(['평균'], as_index=False).mean())
mean = np.round(average.iloc[:, 1:].values,2)
average.iloc[:, 1:] = mean
average = average.rename(index=str, columns={"평균": "역명"})
average
```

```python
ratio = pd.concat([average, ratio])
ratio = ratio.reset_index().drop(['index'], axis=1)
ratio
```

```python
len(ratio)
```

```python
#평균이랑 차이 구하기
def df_dif(dataframe):
    for i in range(len(dataframe)):
        if i > 0:
            dif = dataframe.iloc[i, 1:].values - dataframe.iloc[0, 1:].values
            dataframe.iloc[i, 1:] = dif
    return dataframe

dif = df_dif(ratio)
```

```python
dif
```

```python
#0보다 큰 값(평균보다 큰 값) 구하기
def over_average(x):
    if x is not None:
        if x > 0:
            return x
        else:
            return False
    else:
        return x

for column in dif.iloc[:,1:].columns:
    dif[column] = dif[column].apply(lambda x : over_average(x))
```

```python
dif
```

```python
#2년 연속 평균보다 큰 역 구하기
dif2 = dif.copy()
for index, row in dif2.iterrows():
    for index_c, column in enumerate(dif2.columns):
        if (row[index_c] != False) and (isinstance(row[index_c], str) == False):
            if (index_c == 1) and (row[index_c+1] == False):
                row[index_c] = False
            elif (index_c == len(dif2.columns)-1) and (row[index_c-1] == False):
                row[index_c] = False
            elif (index_c > 1) and (index_c < len(dif2.columns)-1) and (row[index_c - 1] == False) and (row[index_c + 1] == False):
                row[index_c] = False
                
for index, row in dif2.iterrows():
    hasTrue = False
    for index_c, column in enumerate(dif2.columns):
        if (isinstance(row[index_c], str) == False) and (row[index_c] != False):
            hasTrue = True
            break
    
    if hasTrue == False:
        dif2 = dif2.drop(index)
```
