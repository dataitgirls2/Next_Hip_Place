```python
#평균이랑 차이 구하기
def df_dif(dataframe):
    for i in range(len(dataframe)):
        if i > 0:
            dif = dataframe.iloc[i, 1:].values - dataframe.iloc[0, 1:].values
            dataframe.iloc[i, 1:] = dif
    return dataframe

dif = df_dif(ratio)
```

```python
dif
```

```python
#0보다 큰 값(평균보다 큰 값) 구하기
def over_average(x):
    if x is not None:
        if x > 0:
            return x
        else:
            return False
    else:
        return x

for column in dif.iloc[:,1:].columns:
    dif[column] = dif[column].apply(lambda x : over_average(x))
```

```python
dif
```

```python
#2년 연속 평균보다 큰 역 구하기
dif2 = dif.copy()
for index, row in dif2.iterrows():
    for index_c, column in enumerate(dif2.columns):
        if (row[index_c] != False) and (isinstance(row[index_c], str) == False):
            if (index_c == 1) and (row[index_c+1] == False):
                row[index_c] = False
            elif (index_c == len(dif2.columns)-1) and (row[index_c-1] == False):
                row[index_c] = False
            elif (index_c > 1) and (index_c < len(dif2.columns)-1) and (row[index_c - 1] == False) and (row[index_c + 1] == False):
                row[index_c] = False
                
for index, row in dif2.iterrows():
    hasTrue = False
    for index_c, column in enumerate(dif2.columns):
        if (isinstance(row[index_c], str) == False) and (row[index_c] != False):
            hasTrue = True
            break
    
    if hasTrue == False:
        dif2 = dif2.drop(index)
```
