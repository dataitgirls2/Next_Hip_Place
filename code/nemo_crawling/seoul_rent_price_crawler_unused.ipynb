# 네모 사이트 크롤러

1. 구역을 16개로 나누어 지역과 시세를 가져와 딕셔너리 형태로 반환

2. 반환된 딕셔너리들을 하나의 리스트에 담기

3. 리스트의 내용을 중복없이 다시 정리

4. 서울시 모든 법정동이 담긴 데이터 프레임에 머지하기 (서울시가 아닌 동이 모두 삭제됨)

```python
def crawler(year, row, col, pivot_lng = 126.6533193221436, pivot_lat = 37.3660954, lng = 0.1615333557129, lat = 0.09439729101419):
    driver = webdriver.Chrome('C:/Users/rlath/Downloads/chromedriver')
    self_lng = pivot_lng + col*lng
    self_lat = pivot_lat + row*lat
    url = f'https://www.nemoapp.kr/Search/MarketSalePrice?DealYear={year}&BuildingCategory=1&SWLng={self_lng}&SWLat={self_lat}&NELng={self_lng+lng}&NELat={self_lat+lat}&Zoom=13'
    driver.get(url)
    driver.maximize_window()
    time.sleep(2)

    html = driver.page_source
    bs = BeautifulSoup(html, 'html5lib')
    htmls = bs.select('.market-price-region-cluster')
    areas = [area.text for area in htmls]

    regex = r'([0-9|\,]+$)'
    prices = []
    for index, area in enumerate(areas):
        price = re.search(regex, area)
        price = price.group(0)
        prices.append(price)
        areas[index] = area.replace(price, '')

    return dict(zip(areas, prices))
```

```python
area_code = pd.read_csv('data/seoul_official_areas_with_code.csv')
area_code = area_code.drop('Unnamed: 0', axis=1)
area_code['법정동명'] = area_code['법정동명'].str.extract(r'\w+구\s([\w\d]+[동가로])')
area_code = area_code[['법정동코드', '법정동명']]
area_code.head()
```

```python
df_result = area_code.copy(deep=True)
df_result.head()
```

```python
crawler(2018, 2, 2)
```

```python
# 한 번에 다 돌리기
for year in range(2012, 2019):
    result = []
    for r in range(4):
        for c in range(4):
            result.append(crawler(year, r, c))

    result_dict = {}
    for dic in result:
        for k, v in dic.items():
            if k not in result_dict.keys():
                result_dict[k] = int(v.replace(',', ''))

    data = {'법정동명' : list(result_dict.keys()), year : list(result_dict.values())}
    crawled = pd.DataFrame.from_dict(data)
    df_result = df_result.merge(crawled, how='left')
df_result.head()
```
