```python
driver = webdriver.Chrome('C:/Users/rlath/Downloads/chromedriver')

year = 2018
url_01 = f'https://www.nemoapp.kr/Search/MarketSalePrice?DealYear={year}&BuildingCategory=1&SWLng=126.86381102191751&SWLat=37.545331577814544&NELng=127.0253443776304&NELat=37.63972886882873&Zoom=13'
driver.get(url_01)
driver.maximize_window()
time.sleep(2)

html = driver.page_source
bs = BeautifulSoup(html, 'html5lib')

htmls = bs.select('.market-price-region-cluster')
areas_2018_01 = [area.text for area in htmls]

regex = r'([0-9|\,]+$)'

prices_2018_01 = []
for index, area in enumerate(areas_2018_01):
    price = re.search(regex, area)
    price = price.group(0)
    prices_2018_01.append(price)
    areas_2018_01[index] = area.replace(price, '')

prices_2018_01, areas_2018_01
```
