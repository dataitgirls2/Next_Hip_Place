```python
def trans_int(dataframe):
    col = dataframe.columns[4:]
    for day in col:
        dataframe[day] = dataframe[day].str.replace(r'\D+', '').astype(int)
    return dataframe

def get_year(lst):
    table = pd.DataFrame(columns = ['역명'])
    for year in lst:
        year = str(year)
        route = "data/" + year + ".csv"
        if year == '2008' :  encoding = 'utf-8'
        else: encoding = 'cp949'
        df = pd.read_csv(route, encoding=encoding)
        
        df = df.loc[df['구분'] == '하차']
        df = trans_int(df)
        df['역명'] = df['역명'].str.split("(").str[0]
        df = pd.DataFrame(df.groupby(['역명'], as_index=False).sum())
        
        df[year] = 0
        for i in range(20):
            i = i+2
            df[year] += df.iloc[:,i]
        df = df[['역명', year]]
        table = pd.merge(table, df, how = 'outer')
    return table
```

```python
lst = [2008, 2009, 2010, 2011, 2012, 2013, 2014]

get_year(lst)
```

```python
get_year(lst).to_csv('data/2008-2014.csv', index=False)
```
