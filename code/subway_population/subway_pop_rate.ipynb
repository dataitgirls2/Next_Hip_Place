```python
import pandas as pd
```

```python
import numpy as np
```

```python
def get_year_data_2015(year, region):
    table = pd.DataFrame(columns = ['역명', year])
    lst = list(map(lambda x: x+1, range(12)))
    
    for month in lst:
        if month < 10:
            route = "data/" + year + "0" + str(month) + ".csv"
        else:
            route = "data/" + year + str(month) + ".csv"
        
        if str(month) == '8' :  encoding = 'utf-8'
        else: encoding = 'cp949'
            
        df = pd.read_csv(route, encoding=encoding)
        
        df['역명'] = df['역명'].astype(str)
        df = df[['역명', '하차총승객수']]
        df = df.loc[df['역명'].isin(region)]
        df = df.groupby('역명').sum().reset_index()
        table = pd.merge(table, df, how='outer')
        table.loc[table['하차총승객수'].isnull(), '하차총승객수'] = 0
        table.loc[table[year].isnull(), year] = 0
        table[year] += table['하차총승객수']
        table = table.drop(['하차총승객수'], axis=1)
    return table
```

```python
def get_year_data_2016(year, region):
    table = pd.DataFrame(columns = ['역명', year])
    lst = list(map(lambda x: x+1, range(12)))
    encoding = 'cp949'
    for month in lst:
        if month < 10:
            route = "data/" + year + "0" + str(month) + ".csv"
        else:
            route = "data/" + year + str(month) + ".csv"
            
        df = pd.read_csv(route, encoding=encoding)
        
        df['역명'] = df['역명'].astype(str)
        df = df[['역명', '하차총승객수']]
        df = df.loc[df['역명'].isin(region)]
        df = df.groupby('역명').sum().reset_index()
        table = pd.merge(table, df, how='outer')
        table.loc[table['하차총승객수'].isnull(), '하차총승객수'] = 0
        table.loc[table[year].isnull(), year] = 0
        table[year] += table['하차총승객수']
        table = table.drop(['하차총승객수'], axis=1)
    return table
```

```python
def get_year_data_2017(year, region):
    table = pd.DataFrame(columns = ['역명', year])
    lst = list(map(lambda x: x+1, range(12)))
    
    for month in lst:
        if month < 10:
            route = "data/" + year + "0" + str(month) + ".csv"
        else:
            route = "data/" + year + str(month) + ".csv"
        if month < 3 :  encoding = 'cp949'
        else: encoding = 'utf-8'
            
        df = pd.read_csv(route, encoding=encoding)      
        df['역명'] = df['역명'].astype(str)
        df = df[['역명', '하차총승객수']]
        df = df.loc[df['역명'].isin(region)]
        df = df.groupby('역명').sum().reset_index()
        table = pd.merge(table, df, how='outer')
        table.loc[table['하차총승객수'].isnull(), '하차총승객수'] = 0
        table.loc[table[year].isnull(), year] = 0
        table[year] += table['하차총승객수']
        table = table.drop(['하차총승객수'], axis=1)
    return table
```

```python
def get_year_data_2018(year, region):
    table = pd.DataFrame(columns = ['역명', year])
    lst = list(map(lambda x: x+1, range(7)))
    encoding = 'utf-8'
    for month in lst:
        route = "data/" + year + "0" + str(month) + ".csv"      
        df = pd.read_csv(route, encoding=encoding)

        df = df[['역ID', '승차총승객수']]
        
        df = df.loc[df['역ID'].isin(region)]
        df = df.groupby('역ID').sum().reset_index()
        df.columns = ['역명', '하차총승객수']
        table = pd.merge(table, df, how='outer')
        table.loc[table['하차총승객수'].isnull(), '하차총승객수'] = 0
        table.loc[table[year].isnull(), year] = 0
        table[year] += table['하차총승객수']
        table = table.drop(['하차총승객수'], axis=1)
    return table
```

```python
#2008-2014년 전처리한 데이터
sub = pd.read_csv('data/2008-2014.csv')
```

```python
sub.head()
```

```python
region = ['홍대입구', '종로3가', '성수', '낙성대', '서울대입구', '망원', '합정', '신사', '안국', '명동', '상도', '오목교', '상계']

sub = sub.loc[sub['역명'].isin(region)]
```

```python
sub
```

```python
df = pd.merge(sub, get_year_data_2015('2015', region), on='역명')
df
```

```python
df = pd.merge(df, get_year_data_2016('2016', region), on='역명')
df
```

```python
df = pd.merge(df, get_year_data_2017('2017', region), on='역명')
df
```

```python
df = pd.merge(df, get_year_data_2018('2018', region), on='역명')
df
```

## 증가율 구하기

```python
before_sub = df.drop(['역명', '2017', '2018'], axis=1)
before_sub.head()
```

```python
after_sub = df.drop(['역명', '2008', '2018'], axis=1)
after_sub.head()
```

```python
increase_rate = pd.DataFrame(np.round(((after_sub.values - before_sub.values)/ before_sub.values)*100, 3))
col_list = [str(num)+'~'+str(num+1) for num in range(2008, 2017)]
increase_rate.columns = col_list
```

```python
sub_rate = pd.DataFrame()
sub_rate['역명'] = df['역명'].copy(deep=True)
sub_rate = sub_rate.join(increase_rate)
sub_rate
```
