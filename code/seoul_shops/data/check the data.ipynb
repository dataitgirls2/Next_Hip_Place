```python
import pandas as pd
from plotnine import *
import re
import numpy as np
```

```python
raw_data = pd.read_csv('shops_original.csv')
raw_data.head()
```

```python
raw_seoul = raw_data.copy(deep=True)
raw_seoul = raw_seoul.loc[(raw_seoul['address'].str.contains('서울특별시'))]
print(raw_seoul.shape)
raw_seoul.head()
```

```python
raw_seoul.loc[raw_seoul['address'].str.contains(r'(\w+[0-9]가)')]
```

```python
raw_seoul['area'] = raw_seoul['address'].str.extract(r'([가-힣|0-9]+[동|가])[^종로구]')
raw_seoul['area'][165426] = '개봉동'
print(raw_seoul.shape)
raw_seoul.head()
```

```python
raw_seoul['area'].loc[raw_seoul['area'].str.contains(r'[0-9]동$') == True].unique()
```

```python
raw_seoul.loc[raw_seoul['area'].str.contains(r'[^상도1동][0-9]+동') == True]
```

```python
for i in raw_seoul.index:
    if raw_seoul.iloc[i]['area'] == r'성수동[12]가':
        raw_seoul.iloc[i]['area'] = '성수동'
    elif raw_seoul.iloc[i]['area'] == r'명동[12]가':
        raw_seoul.iloc[i]['area'] = '명동'
    elif raw_seoul.iloc[i]['area'] == r'상도1*동':
        raw_seoul.iloc[i]['area'] = '상도동'
```

```python
areas = []
areas = [area for area in raw_seoul['area'] if area not in areas]
```

```python
areas = []
for area in raw_seoul['area']:
    if area not in areas:
        areas.append(area)
len(areas)
```

```python
seoul = pd.DataFrame()
seoul['area'] = areas
seoul.head()
```
