## 서울 시내 요식업체 수

```python
raw_seoul = raw_data.copy(deep=True)
raw_seoul = raw_seoul.loc[(raw_seoul['address'].str.contains('서울특별시'))]
print(raw_seoul.shape)
raw_seoul.head()
```

```python
raw_seoul['area'] = raw_seoul['address'].str.extract(r'([가-힣|0-9]+[동|가])[^종로구]')
raw_seoul['area'][165426] = '개봉동'
print(raw_seoul.shape)
raw_seoul.head()
```

# 연도별 운영 점포 순위

```python
def seoul_year(df, year, result):
    df_year = df.loc[(df['영업자시작년도'] <= year) & ((df['폐업년도'] > year) | (df['폐업년도'].isnull()))]

    area_counts = {}
    for area in df_year['area']:
        if area not in area_counts.keys():
            area_counts[area] = 1
        else:
            area_counts[area] += 1
            
    df_result = pd.DataFrame()
    df_result['area'] = area_counts.keys()
    df_result[year] = area_counts.values()

    result = result.merge(df_result, on='area')

    return result
```

```python
seoul_area = {}

for area in raw_seoul['area']:
    if area not in seoul_area.keys():
        seoul_area[area] = 1
    else:
        seoul_area[area] += 1
seoul_area
```

```python
# count = 0
# for k, v in seoul_area.items():
#     if v <= 15:
#         print(k, ':', v)
#         count += 1
# print(f'\n\n{count}')
```

```python
seoul = pd.DataFrame()
seoul['area'] = seoul_area.keys()
seoul.head()
```

```python
# 연도별 서울 시내 운영점포 수 데이터 프레임 만들기
for year in range(2007, 2019):
    seoul = seoul_year(raw_seoul, year, seoul)
seoul.head()
```

```python
print(seoul.shape)
```

```python
sorted_result = pd.DataFrame()
for year in range(2007, 2019):
    sorted_by_year = seoul.sort_values(by=year, ascending=False)
    area_and_year = [f'{sorted_by_year["area"][i]} ({sorted_by_year[year][i]})'\
                     for i in sorted_by_year.index]
    sorted_result[year] = area_and_year
sorted_result.head()
```

# 연도별 운영 점포 증가율 순위

```python
before_seoul = seoul.drop('area', axis=1)
before_seoul.drop(2018, axis=1, inplace=True)
before_seoul.head()
```

```python
after_seoul = seoul.drop('area', axis=1)
after_seoul.drop(2007, axis=1, inplace=True)
after_seoul.head()
```

```python
import numpy as np
```

```python
seoul_shops_rate = pd.DataFrame(np.round((after_seoul.values - before_seoul.values) * 100 / before_seoul.values, 3))
col_list = [str(num)+'~'+str(num+1) for num in range(2007, 2018)]
seoul_shops_rate.columns = col_list
final_rate = pd.DataFrame()
final_rate['area'] = seoul['area'].copy(deep=True)
final_rate = final_rate.join(seoul_shops_rate)
final_rate.head()
```

```python
sorted_rate_result = pd.DataFrame()
for year in range(2007, 2018):
    col_name = f'{year}~{year+1}'
    sorted_by_year = final_rate.sort_values(by=col_name, ascending=False)
    area_and_year = [f'{sorted_by_year["area"][i]} ({sorted_by_year[col_name][i]})'\
                     for i in sorted_by_year.index]
    sorted_rate_result[col_name] = area_and_year
sorted_rate_result.head()
```

```python
# sorted_rate_result.head(30)
```

```python
sorted_rate_result.to_csv('data/seoul_shops_rate_sorted_by_year.csv', index=False)
```

# 연도별 개업 점포 순위

```python
def new_shops_year(df, year, result):
    df_year = df.loc[(df['영업자시작년도'] == year)]
    
    area_counts = {}
    for area in df_year['area']:
        if area not in area_counts.keys():
            area_counts[area] = 1
        else:
            area_counts[area] += 1
            
    df_result = pd.DataFrame()
    df_result['area'] = area_counts.keys()
    df_result[year] = area_counts.values()

    result = result.merge(df_result, on='area')

    return result
```

```python
opened_shops = pd.DataFrame()
opened_shops['area'] = seoul_area.keys()

# 연도별 서울 시내 개업수 데이터 프레임 만들기
for year in range(2007, 2019):
    opened_shops = new_shops_year(raw_seoul, year, opened_shops)
opened_shops.head()
```

```python
sorted_result_opened = pd.DataFrame()
for year in range(2007, 2019):
    sorted_by_year = opened_shops.sort_values(by=year, ascending=False)
    area_and_year = [f'{sorted_by_year["area"][i]} ({sorted_by_year[year][i]})'\
                     for i in sorted_by_year.index]
    sorted_result_opened[year] = area_and_year
sorted_result_opened.head()
```

```python
sorted_result_opened.to_csv('data/seoul_opened_shops_sorted_by_year.csv', index=False)
```
