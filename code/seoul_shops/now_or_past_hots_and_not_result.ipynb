```python
import pandas as pd
from plotnine import *
import re
import numpy as np
```

```python
def area_rank(df, area):
    ranks = [df.loc[df[str(year)].str.contains(area)].index[0] for year in range(2007, 2019)]
    return ranks
```

```python
now_hot = '연남동,연희동,익선동,성수동,삼성동,망원동,봉천동'.split(',')
past_hot = '신사동,삼청동,명동'.split(',')
not_hot = '상도동,목동,신정동,상계동'.split(',')
```

```python
sorted_shops_abs = pd.read_csv('data/seoul_shops_sorted_by_year.csv')
sorted_shops_rate = pd.read_csv('data/seoul_shops_rate_sorted_by_year.csv')
sorted_opened_abs = pd.read_csv('data/seoul_opened_shops_sorted_by_year.csv')
sorted_opened_rate = pd.read_csv('data/seoul_opened_shops_rate_sorted_by_year.csv')
```

```python
df_now_null = pd.DataFrame(index=now_hot)
for year in range(2007, 2019):
    df_now_null[year] = np.nan
```

```python
now_hot_df = df_now_null.copy(deep=True)
i = 0
for area in now_hot:
    ranks = area_rank(sorted_shops_abs, area)
    values = dict(zip(range(2007, 2019), ranks))
    now_hot_df.iloc[i] = now_hot_df.iloc[i].fillna(value=values, axis=0)
    i += 1
now_hot_df = now_hot_df.astype(int)

now_hot_df.to_csv('data/now_hot_place_rank_sorted_shops_abs.csv', index=False)
```
