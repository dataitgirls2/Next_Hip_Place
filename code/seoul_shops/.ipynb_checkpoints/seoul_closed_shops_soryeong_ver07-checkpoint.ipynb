```python
import pandas as pd
from plotnine import *
import re
```

```python
df = pd.read_csv('data/data_raw_seoul_shops.csv')
```

```python
df.head()
```

```python
df_seoul = df.copy(deep=True)
```

```python
df_seoul = df_seoul.ix[(df_seoul['address'].str.contains('서울특별시'))]
```

```python
df_seoul['area'] = df_seoul['address'].str.extract(r'([가-힣|0-9]+[동|가])[^종로구]')
```

```python
df_seoul['area'][165426] = '개봉동'
```

```python
def seoul_year(df_seoul, year, seoul):
    df_year = df_seoul.loc[(df_seoul['영업자시작년도'] <= year) & ((df_seoul['폐업년도'] > year) | (df_seoul['폐업년도'].isnull()))]
#     df_year = df_year.drop(['address', 'lat', 'lng', '구', '년도', '업소명', '업종명', '업태명'], axis=1)

    area_counts = {}
    for area in df_year['area']:
        if area not in area_counts.keys():
            area_counts[area] = 1
        else:
            area_counts[area] += 1
            
    df_result = pd.DataFrame()
    df_result['area'] = area_counts.keys()
    df_result[year] = area_counts.values()

    seoul = seoul.merge(df_result, on='area')

    return seoul
```

```python
seoul_area = {}

for area in df_seoul['area']:
    if area not in seoul_area.keys():
        seoul_area[area] = 1
    else:
        seoul_area[area] += 1
seoul_area
```

```python
seoul = pd.DataFrame()
```

```python
seoul['area'] = seoul_area.keys()
```

```python
seoul.head()
```

```python
for year in range(2007, 2019):
    seoul = seoul_year(df_seoul, year, seoul)
seoul.head()
```

```python
seoul.to_csv('data/data_seoul_shops.csv', index=False)
```
