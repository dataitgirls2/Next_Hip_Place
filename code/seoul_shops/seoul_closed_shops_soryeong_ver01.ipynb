```python
def opened_at(df, year, area):
    df_year = df.loc[df['영업자시작년도'] == year]
    df_area = df_year[df_year['address'].str.contains(area)]
    
    return df_area
```

```python
def closed_at(df, year, area):
    df_year = df.loc[df['폐업년도'] == year]
    df_area = df_year[df_year['address'].str.contains(area)]
    
    return df_area
```

```python
def where_opened_or_closed(df, area, year=None, status='opened'):
    df_area = df[df['address'].str.contains(area)]
    
    if status == 'opened':
        df_year = df_area.loc[df_area['영업자시작년도'] == year]
    elif status == 'closed':
        df_year = df_area.loc[df_area['폐업년도'] == year]
    else:
        return 0
    
    return df_area
```

```python
def searching(df, area, year=None, status=None):
    df_area = df[df['address'].str.contains(area)]
    if year != None:
        if status == 'opened':
            df_area = df_area.loc[df_area['영업자시작년도'] == year]
        elif status == 'closed':
            df_area = df_area.loc[df_area['폐업년도'] == year]
        else:
            return 'you missed status(opened or closed)'
    
    return df_area
```
