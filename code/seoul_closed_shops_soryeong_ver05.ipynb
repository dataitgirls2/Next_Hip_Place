```python
def seoul_shops(df, year, result):
    df_year = df.loc[(df['영업자시작년도'] <= year) & ((df['폐업년도'] > year) | (df['폐업년도'].isnull()))]
    
    seoul_area = {}
    for area in df_year['area']:
        if area not in seoul_area.keys():
            seoul_area[area] = 1
        else:
            seoul_area[area] += 1
            
    result2 = result.merge(df_year, on='area')
    
    return result2
```

```python

for year in range(2007, 2008):
    result = seoul_shops(df_seoul, year, seoul)
    
result
```

```python
def seoul_dataframe(df, year):
    df_year = df.loc[(df['영업자시작년도'] <= year) & ((df['폐업년도'] > year) | df['폐업년도'].isnull())]

    seoul_area = {}
    for area in df_year['area']:
        if area not in seoul_area.keys():
            seoul_area[area] = 0
        else:
            seoul_area[area] += 1

    seoul = pd.DataFrame()
    seoul['area'] = seoul_area.keys()
    seoul[str(year)] = seoul_area.values()

    return seoul
```
